```python
import matplotlib.pyplot as plt
```

```python
keys, vals = [], []
with open("../../output_firstletter/part-r-00000", "r") as file:
    for t in file:
        k, v = t.strip().split("\t")
        keys.append(k)
        vals.append(int(v))
```

```python
plt.figure(figsize=(13, 5))
plt.bar(keys, vals)
plt.yscale("log")
plt.grid(color='r', linestyle='-', linewidth=1, axis="y", which="major")
plt.show()
```
