```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("pronoun.txt", sep=" ", header=None)
df.columns = ["pronoun", "count"]
df["proportion"] = df["count"]/sum(df["count"])
df
```

```python
plt.figure(figsize=(13, 5))
plt.bar(df["pronoun"], df["count"])
plt.yscale("log")
plt.xlabel("Freq")
plt.ylabel("Pronoun")
plt.title("The Frequency of Pronoun Present in Tweets")
plt.grid(color='r', linestyle='-', linewidth=1, axis="y", which="major")
plt.show()
```
